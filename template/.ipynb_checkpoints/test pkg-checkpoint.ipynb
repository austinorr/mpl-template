```python
def insert_image(ax, image_path, scale=1, dpi=300, expand = False):
    """
    Centers an image within an axes object

    Parameters
    ----------
    ax : matplotlib.Axes
    image_path : str
        Path to an existing image file.
    scale : float, optional
        The relative scale of the desired output. 
        Values should be positive floats. 
        E.g. scale = 2 will double the image's size
        relative to the given matplotlib.Axes object.
        scale = 0.5 will scale the image to half of the 
        given matplotlib.Axes object.
        
    dpi : int, optonal (default = 300)
        The Dots (pixel) Per Inch of the image.

    Notes
    -----
    Use scale parameter to zoom image relative to axes boundary.
    """
    imgaxes = ax.figure.add_axes(ax.get_position(), xticks=[], yticks=[], zorder = 1)
    bbox = ax.get_window_extent().transformed(ax.get_figure().dpi_scale_trans.inverted())
    width, height = bbox.width, bbox.height
    width *= dpi
    height *= dpi
    img = _validate_image_path(image_path)

    with Image.open(img) as image:
        image = image.convert('RGBA')
        wpx, hpx = image.size
        as_im = wpx/hpx
        as_ax = width/height
        if scale > 1:
            #if expand:
                #expand and zoom
            #else:
                #zoom in place
            if (as_im < as_ax):
                adjy = (hpx/scale)/2
                if expand:
                    adjx = (width/height)*adjy #adjust width, keep height
                else:
                    adjx = (wpx/scale)/2
            else:
                adjx = (wpx/scale)/2
                if expand:
                    adjy = (height/width)*adjx
                else:
                    adjy = (hpx/scale)/2
            image = image.crop(
                (int(wpx/2-adjx),
                 int(hpx/2-adjy),
                 int(wpx/2+adjx),
                 int(hpx/2+adjy)
                 )
                )

            image = image.resize(image.size, Image.BICUBIC)
            
        else:
            if width >= height:
                width = int(wpx * (height/hpx))
            else:
                height = int(hpx * (width/wpx))
            image = image.resize((int(width*scale), int(height*scale)), Image.ANTIALIAS)
            # image = np.asarray(image)
            # print(image)

            imgaxes.set_xlim(_calc_extents(image.shape[1], scale))
            imgaxes.set_ylim(reversed(_calc_extents(image.shape[0], scale)))

    imgaxes.imshow(image, aspect='equal')

    return imgaxes
```
