```python
user = [{'name':'Title', #use as ax label
         'text':[
            {
                "ha":"center",
                "s":"Geosyntec Template Example Plots",
                "va":"baseline",
                "weight":"bold",
                "x":0.5,
                "y":0.52,
                },
            {
                "color":(.3,0.3,0.3),
                "ha":"center",
                "s":"Blank Example",
                "va":"top",
                "weight":"light",
                "x":0.5,
                "y":0.48,
                },
            ],
         },
        {'name':'Logo',
         'image':{'path':'img//geosyntec-logo.png',
                  'scale':.8,
                  },
         },
        {'name':'Project',
         'text':{
                "ha":"center",
                "s":"WR####",
                "va":"center",
                "x":0.5,
                "y":0.5,

            },
         },
        {'name':'Date',
         'text':{

                    "ha":"center",
                    "s":"May 2016",
                    "va":"center",
                    "x":0.5,
                    "y":0.5,
                },

         },
        {'name':'Fignum',
         'text':{
                "ha":"center",
                "s":"Figure\n\n{:02d}",
                "va":"center",
                "weight":"bold",
                "x":0.5,
                "y":0.5,
                },

         },
        ]

# TBK =   [{'name':   "Title",
#           'text':{
#                 "title":{
#                     "ha":"center",
#                     "s":"Geosyntec Template Example Plots",
#                     "va":"baseline",
#                     "weight":"bold",
#                     "x":0.5,
#                     "y":0.52
#                 },
#                 "subtitle":{
#                     "color":(.3,0.3,0.3),
#                     "ha":"center",
#                     "s":"Blank Example",
#                     "va":"top",
#                     "weight":"light",
#                     "x":0.5,
#                     "y":0.48
#                 },
# #         "span":{},
#             },
#           },
#          {'name':
#             "b_3":{
#                 "project":{
#                     "ha":"center",
#                     "s":"WR####",
#                     "va":"center",
#                     "x":0.5,
#                     "y":0.5
#                 }
#             },
#             "b_4":{
#                 "date":{
#                     "ha":"center",
#                     "s":"May 2016",
#                     "va":"center",
#                     "x":0.5,
#                     "y":0.5
#                 }
#             },
#             "b_5":{
#                 "fignum":{
#                     "ha":"center",
#                     "s":"Figure\n\n{:02d}",
#                     "va":"center",
#                     "weight":"bold",
#                     "x":0.5,
#                     "y":0.5
#                 }
#             }


def add_tbk_text(ax, tbk_elem): # add logo ax label name, specify path, add function to source?
    """
    This is a custom titleblock handler, will need to be updated for 
    any new projects to catch special cases.
    
    ax: mpl.axes() object
    tbk_elem: dictionary to populate titleblock contents. Must use 
        mpl.pyplot.text() object keywords
    
    """
    logo = "http://matplotlib.org/_images/polar_bar_demo.png" # "img//geosyntec-logo.png" #test @ url maybe need urllib
    
    label = ax.get_label() # the GeosyntecTemplate class creates labels
                           # b_1 through b_5 for populating the titleblock

    if label=='b_2':
        logo_ax = template.insert_image(ax, logo, scale = .8, dpi = ax.get_figure().get_dpi())
        logo_ax.set_label('logo')
        logo_ax.axis('off')
    
    elif label in tbk_elem.keys():
        for i in tbk_elem[label]:
            if i != 'span': # span is an advanced function that we'll talk about later. It controls how many columns and rows
                kwargs = copy.deepcopy(tbk_elem[label][i])
                ax.text(**kwargs)
```

```python
def _spans(rows, cols):
    spans = [{'span': [0, rows[0], 0,sum(cols)]},
             {'span': [rows[0], rows[0]+rows[1], 0, cols[0]+cols[1]]},
             {'span': [rows[0]+rows[1], sum(rows), 0, cols[0]]},
             {'span': [rows[0]+rows[1], sum(rows), cols[1], cols[0]+cols[1]]},
             {'span': [rows[0], sum(rows), cols[0]+cols[1], sum(cols)]},
             ]
    return spans
```
